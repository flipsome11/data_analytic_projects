```python
import yfinance as yf

```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
#downloaded sp500 data
sp500
```

```python
#clean and visualize the data
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500['Dividends']
del sp500['Stock Splits']
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
#setup our machine learning target
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
#training initial model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]
# we want the model to learn how to predict the stock

predictors = ["Close", "Volume", "Open", "High", "Low"]
# train the model by using predictors column in order to predict the target
model.fit(train[predictors], train["Target"])
```

```python
# next step is measure how accurate the model is

from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
#plot our predictions, combined our actual values with predicted values
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
# orange line, zero is our prediction, blue line is wut actually happened
```

```python
#enable back testing, create a prediction function

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
#write a backtest function, have a certain amount of data to train model, 10yrs of data
#this back testing will allow more confidence in our model as it accumulate data yearly to predict next year
#evaluae error and create a way for backtest, accurately measure error over period of time
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
        
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
# see how many days we predicted the value go up/down
predictions["Predictions"].value_counts()
```

```python
#give us our precision score, about 53 %accurate, is this gd?
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
#as a benchmark, we can compare wif the percentage the market went up

predictions["Target"].value_counts() / predictions.shape[0]

#s&p500 went up 53% of days but went down 46% on days
```

```python
# create a variety of rolling averages, the inputs that determine stock go up/down, giving the algorithm that info
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_(horizon)"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
#a trend of the horizon days that stock prices went up
    trend_column = f"Trend_(horizon)"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
#define what becomes a 1 or a zero
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
#more days the price went up after we asked the model to be more confident in its predictions
#this means buying stocks on fewer days
predictions["Predictions"].value_counts()
```

```python
# when the model predict price goes up, 52% of the time it will actually go up, 
# based on time series data and historical prices of the index
precision_score(predictions["Target"], predictions['Predictions'])
```
